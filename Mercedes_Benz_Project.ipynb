```python
#Importing Libraries
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
#pd.options.display.float_format = '{:,.4f}'.format
variance = data.var()
variance = variance.reset_index() #we do this so that the variance is displayed as columns
variance.columns = ["id","values"] #variable name=ID values= variance of each column 
variance= variance.sort_values("values",ascending=1)#we sort the variance column having index 1 in asceding order
variance
```

```python
# We will remove the varialbes with variance 0 and low variance (less than 0.2)--as it lacks variability of data,and
#information stored by these variables are less too.
#extremely large variance needs to be dropped as well bcz it can dominate other variances
# We will also remove id since it has a huge variance--these variables will dominate and nullify the impact of other variables
var = variance.loc[variance["values"] < 0.2,"id"]
data1 = data.drop(var,axis=1)
data1.drop("ID",axis=1,inplace=True)#inplace = true means we are making permanent fix
data1.head()
```

```python
data.shape
```

```python
# Missing values
data1[data1.isnull().any(axis=1)]
# We don't have any missing values in the data
```

```python
c = data1.corr().abs()
s = c.unstack() #does transpose rows to columns
s = pd.DataFrame(s)
s.reset_index(inplace = True)
s.head()
```

```python
import numpy as np
s["flag"] = np.where(s["level_0"] == s["level_1"],"same","not same")
s.columns.values[2] = "corr"
s.head()
```

```python
# Remove the variables with correlation more than .9
#s.loc[s["flag"] != "same",]
name = s.loc[(s["corr"] > .9) & (s["flag"] != "same") ,"level_1"]
final_name = name.unique()
```

```python
data2 = data1.drop(final_name,axis=1)
data2.head()
```

```python
data2.info()
```

### Creating Dummies Variables

```python
char = data2.select_dtypes(exclude='number')
num = data2.select_dtypes(include='number')
```

```python
char1 = pd.get_dummies(char.astype(str),drop_first=True)
char1.head()
```

```python
data3 = pd.concat([char1,num],axis=1)
data3.head()
```

### XGBOOST Model to calculate the mape
#### MAPE(Mean Absolute Percentage Error)

```python
#segregating ind and dep var
X = data3.drop("y",axis=1)
y = data3.loc[:,"y"]
# Splitting
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y,test_size = 0.3, random_state = 123)
# Instantiation
import xgboost as xg
xgb_r = xg.XGBRegressor(objective ='reg:squarederror',n_estimators = 10, seed = 123)
# Fitting the model
xgb_r.fit(train_X, train_y)
# Predict the model
y_pred = xgb_r.predict(test_X)
###### mape
d = pd.DataFrame() #creating an empty dataframe
d["test_y"] = test_y
d["y_pred"] = y_pred
d["mp"] = abs((d["test_y"]- d["y_pred"])/d["test_y"])
(d.mp.mean())*100#mape
```

### Lower error component helps to increase accuracy of the model
